```python
# select necessary columns
select_columns = ['SKU', 'Item ID', 'Publish Status', 'Product Category', 'Price', 'Buy Box Item Price', 'MSRP', 'Shipping Weight', 'WPID', 'GTIN', 'Shelf Name', 'Primary Category Path', 'Brand', 'Offer Start Date', 'Offer End Date', 'Item Creation Date', 'Item Last Updated', 'Reviews Count', 'Average Rating', 'Competitor Price', 'Competitor Ship Price', 'Competitor Last Date Fetched', 'Condition', 'Product Type', 'Impressions', 'Clicks', 'Added to Cart', 'Sales Rank', 'Transactability Score', 'Buy Box Win %']
select_data = data.loc[:, select_columns]
select_data.shape
select_data.head()
```

```python
select_data['Primary Category Path'] = select_data['Primary Category Path'].str.replace(r'["\[\]]', '', regex = True)
select_data['Primary Category Path'] = select_data['Primary Category Path'].str.split(',')
select_data['Primary Category Path'] = select_data['Primary Category Path'].apply(tuple)
select_data['Primary Category Path'].head()
```

```python
select_data.fillna({'Average Rating': 0}, inplace = True)
select_data.fillna({'MSRP': -1}, inplace = True) # marked as unknown
select_data.loc[select_data['Sales Rank'] == '-', 'Sales Rank'] = -1 # unknow
select_data['Sales Rank'] = select_data['Sales Rank'].astype('int')
select_data.dropna(inplace=True)
print(f'Duplicated rows: {select_data.duplicated().sum()}')
select_data.drop_duplicates(inplace=True)
select_data.shape
```
